# 一、Matplotlib 学习

## 1.1 基础绘图学习

```python
## 学习Matplotlib——函数式绘图
# a strait line: use pyplot functionsfrom matplotlib.pyplot import *
import matplotlib.pyplot as plt
plt.plot([0, 1], [0, 1])      # plot a line from (0, 0) to (1, 1)
plt.title("a strait line")
plt.xlabel("x value")
plt.ylabel("y value")
plt.show()
```

```python
#1D data
x = [1,2,3,4,5]
y = [2.3,3.4,4.5,5.6,7.0]
plt.figure(figsize = (12,6))

plt.subplot(231)
plt.plot(x,y)
plt.title("plot")

plt.subplot(232)
plt.scatter(x,y)
plt.title("scatter")

plt.subplot(233)
plt.pie(y)
plt.title("pie")

plt.subplot(234)
plt.bar(x,y)
plt.title("bar")

#2D data
import numpy as np 
delta = 0.025
x = y = np.arange(-3.0,3.0,delta)
X, Y = np.meshgrid(x,y)
Z = Y**2+X**2

plt.subplot(235)
plt.contour(X,Y,Z)
plt.colorbar()
plt.title("contour")

#read image
import matplotlib.image as mpimg
img = mpimg.imread('demo.jpg')

plt.subplot(236)
plt.imshow(img)
plt.title("imshow")

plt.savefig("matplot_sample.jpg")
```

```python
## matplot 中subplot和 subplots的用法
# 1. 如果只画一张图，使用plt.figure
import matplotlib.pyplot as plt

x = [1,2,3,4,5]
y = [2.3,3.4,4.5,5.6,7.0]

plt.figure(figsize=(4,4),dpi=90) #设置画布大小及像素
# plt.xticks([1,5]) #设置x坐标刻度
# plt.yticks([1,10]) #设置y坐标刻度
plt.xlabel("x_test") #设置x坐标名
plt.ylabel("y_test") #设置y坐标名
plt.title("plot_test") #设置主题
plt.plot(x,y) #画图
plt.legend(['blue'])
# plt.axis('off')

```
